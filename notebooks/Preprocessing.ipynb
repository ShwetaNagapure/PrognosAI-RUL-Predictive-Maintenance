```python
## Adding RLU coulmns for preprocessing
def add_rul(df):
    # Find max cycle for each engine in each dataset
    max_cycle = df.groupby(["dataset", "id"])["cycle"].max().reset_index()
    max_cycle.columns = ["dataset", "id", "max_cycle"]

    # Merge max_cycle back into main df
    df = df.merge(max_cycle, on=["dataset", "id"], how="left")

    # RUL = max_cycle - current cycle
    df["RUL"] = df["max_cycle"] - df["cycle"]
    
    return df
```

```python
df = add_rul(df)
```

```python
## Sensor selection Code 

sensor_cols = [col for col in df.columns if "sensor" in col]

print("All sensor columns:", sensor_cols)


variance = df[sensor_cols].var()
low_variance_sensors = variance[variance < 1e-3].index.tolist() ## Sensor with low constant or nearly constant variance

print("Low-variance (remove these):", low_variance_sensors)

useful_sensors = [s for s in sensor_cols if s not in low_variance_sensors]

print("Sensors remaining after variance filter:", useful_sensors)


```

```python
corr = df[useful_sensors].corr()

threshold = 0.99  

corr_matrix = corr.abs()
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

redundant_sensors = [column for column in upper_tri.columns 
                     if any(upper_tri[column] > threshold)]

print("Highly correlated redundant sensors:", redundant_sensors)
```

```python
# Final sensor list
final_sensors = [s for s in useful_sensors if s not in redundant_sensors]

print("Final selected sensors:", final_sensors)
```

```python
import numpy as np

def create_sequences(df, window_size, feature_cols, target_col="RUL"):
    sequences = []
    targets = []

    for unit_id in df['id'].unique():  
        unit_data = df[df['id'] == unit_id].reset_index(drop=True)

        feature_matrix = unit_data[feature_cols].values
        target_values = unit_data[target_col].values

        # rolling window
        for i in range(len(unit_data) - window_size + 1):
            seq = feature_matrix[i : i + window_size]   # X sequence
            target = target_values[i + window_size - 1] # y value at end of window
            sequences.append(seq)
            targets.append(target)

    return np.array(sequences), np.array(targets)
```
